```python
import numpy as np;
from load_dataset import load_mnist_test, load_mnist_train;
from sporco import plot;
from visualize import plot_2;
```

```python
# 検証用画像の読み込み

opt = make_option();
img, _ = load_mnist_train(1);
img = img.squeeze();
plot.imview(img);
```

```python
# 左上寄せ、右下寄せ画像のパワースペクトル画像の表示

left_upper = np.zeros((64, 64));
right_under = np.zeros((64, 64));

left_upper[0:32, 0:32] = img;
# 2次元FFT
img_fft = np.fft.fft2(left_upper);
# 象限の入れ替え。
img_fft =  np.fft.fftshift(img_fft);
# パワースペクトル
img_power1 = 20*np.log(np.abs(img_fft));

plot_2([left_upper, img_power1], ["original image", "power spectrum"]);


right_under[32:64, 32:64] = img;
# 2次元FFT
img_fft = np.fft.fft2(right_under);
# 象限の入れ替え。
img_fft =  np.fft.fftshift(img_fft);
# パワースペクトル
img_power2 = 20*np.log(np.abs(img_fft));

plot_2([right_under, img_power2], ["original image", "power spectrum"]);
```

```python
# パワースペクトルは同じ
plot.imview(np.abs(img_power1-img_power2));
```
