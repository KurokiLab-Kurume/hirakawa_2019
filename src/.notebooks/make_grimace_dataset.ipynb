```python
import numpy as np;
import cv2;
import glob;
import os;
```

```python
# データセット特有の処理
# dahさんとjerさんは特有の処理が必要
path = "./grimace/";
files = os.listdir(path);
files_dir = [path + f for f in files if os.path.isdir(os.path.join(path, f))];
print(files_dir);

for i in files:
    folder_path = path + i + "/";
    prefix = folder_path + i + "_exp.";
    if i == "dah":
        prefix = folder_path + i + "_exp_2.";
    if i == "jer":
        prefix = folder_path + i + "_2.";
    for j in range(0, 20):
        before = prefix + str(j+1) + ".jpg";
        os.rename(before, folder_path + str(j) + ".jpg");

cnt = 0;
for k in files_dir:
    os.rename(k, path + str(cnt));
    cnt = cnt + 1;
```

```python
# 各クラスのデータセットを作る
# ./grimace/dataset にデータセットを作るスクリプト
for i in range(18):
    data = np.zeros((200, 180, 3, 20));
    for j in range(20):
        img = cv2.imread("./grimace/"+str(i)+"/"+str(j)+".jpg");
        data[:, :, :, j] = img;
    np.save(path + "dataset/" + str(i) + ".npy", np.uint8(data));
```
